```python
data = annotations_data[0]
#{'x': 269.5, 'y': 273, 'width': 301, 'height': 334}
[x,y,w,h]= [269,273, 301,  334]

image_path = PATH+'/'+data['image']

image = cv2.imread(image_path) 
    
# Window name in which image is displayed 
window_name = 'Image'
   
# Center coordinates 
center_coordinates = (269 - w//2, 273-h//2) 
center_coordinates2 = ((269 - w//2)+301, (273-h//2)+334) 
print(center_coordinates)
# Radius of circle 
radius = 20
   
# Blue color in BGR 
color1 = (255, 0, 0) 
color2 = (0, 255, 0) 
   
# Line thickness of 2 px 
thickness = 2
   
# Using cv2.circle() method 
# Draw a circle with blue line borders of thickness of 2 px 
image = cv2.circle(image, center_coordinates, radius, color1, thickness) 
image = cv2.circle(image, center_coordinates2, radius, color2, thickness) 

   
# Displaying the image  
cv2.imshow(window_name, image)  
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
